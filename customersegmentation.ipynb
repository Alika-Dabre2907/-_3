```python
df=df.drop(columns=['Z_CostContact','Z_Revenue'])
```

```python
def get_relationship(row):
    if row['marital_Married'] == 1:
        return 1
    elif row['marital_Together'] == 1:
        return 1
    else:
        return 0
        df['relationship'] = df.apply(get_relationship, axis=1)
df.head()
```

```python
#correcting column name
df.rename(columns={'education_2n Cycle\t':'education_2n Cycle'},inplace=True)
```

```python
# assigning value for education
def get_education(row):
    if row['education_2n Cycle'] == 1:
        return '2'
    elif row['education_Basic'] == 1:
        return '1'
    elif row['education_Graduation'] == 1:
        return '3'
    elif row['education_Master'] == 1:
        return '4'
    elif row['education_PhD'] == 1:
        return '5'
    else:
        return 'Unknown'
        df['education'] = df.apply(get_education, axis=1)
df.head()
```

```python
#importing libraries
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
cols_to_cluster = [
    'Income', 'MntTotal', 'NumWebPurchases', 'NumStorePurchases',
    'NumCatalogPurchases', 'Recency', 'Age', 'Kidhome', 'Teenhome',
    'AcceptedCmpOverall'
]

scaler = StandardScaler()
df_scaled = df.copy()
df_scaled[cols_to_cluster] = scaler.fit_transform(df[cols_to_cluster])
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
df_scaled['Cluster'] = kmeans.fit_predict(df_scaled[cols_to_cluster])
print(df_scaled[['Cluster'] + cols_to_cluster].head())
```

```python
#  mean of each feature
cluster_means = df_scaled.groupby('Cluster')[cols_to_cluster].mean()

#  means for each cluster
print(cluster_means)
```

```python
#average income for each cluster
print(df_scaled['Cluster'].isnull().sum())
print(df_scaled['Cluster'].unique())

df_scaled['Cluster'] = df_scaled['Cluster'].astype(int)

average_income_per_cluster = df_scaled.groupby('Cluster')['Income'].mean()

print(average_income_per_cluster)

plt.figure(figsize=(8, 5))
average_income_per_cluster.plot(kind='bar', color='skyblue')
plt.title('Average Income per Cluster', fontsize=16)
plt.xlabel('Cluster', fontsize=12)
plt.ylabel('Average Income', fontsize=12)
plt.xticks(rotation=0)
plt.show()


```

```python

def get_minimum_education(row):
    if row['education_Basic'] == 1:
        return 1  # Basic education
    elif row['education_2n Cycle'] == 1:
        return 2  # 2nd Cycle (High School)
    elif row['education_Graduation'] == 1:
        return 3  # Graduation
    elif row['education_Master'] == 1:
        return 4  # Master's
    elif row['education_PhD'] == 1:
        return 5  # PhD
    else:
        return 0  # No education

# Apply the function to create a new column for minimum education level
df['MinEducationLevel'] = df.apply(get_minimum_education, axis=1)

# Show the first few rows with the new column
print(df[['MinEducationLevel', 'education_Basic', 'education_2n Cycle', 'education_Graduation', 'education_Master', 'education_PhD']].head())
```

```python
print(df[['MinEducationLevel', 'education_Basic', 'education_2n Cycle', 'education_Graduation', 'education_Master', 'education_PhD']].head())
```

```python
#  average purchase value across categories
purchase_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

# Average purchase value per customer
df['AveragePurchaseValue'] = df[purchase_columns].mean(axis=1)

# overall average purchase value
avg_purchase_value = df['AveragePurchaseValue'].mean()
print(f"Average Purchase Value: {avg_purchase_value}")
```

```python
# Frequency of purchases (sum of all purchase methods)
df['TotalPurchases'] = df['NumWebPurchases'] + df['NumStorePurchases'] + df['NumCatalogPurchases']

#  overall frequency of purchases
avg_purchases = df['TotalPurchases'].mean()
print(f"Average Frequency of Purchases: {avg_purchases}")
```

```python
# Total spending per customer
df['TotalSpend'] = df[purchase_columns].sum(axis=1)

# overall total spend
total_spend = df['TotalSpend'].sum()
print(f"Total Spend of All Customers: {total_spend}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

print(df.columns)


cols_to_cluster = ['Income', 'MntTotal', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts']  # Adjust with actual columns


scaler = StandardScaler()
df_scaled = df.copy()
df_scaled[cols_to_cluster] = scaler.fit_transform(df[cols_to_cluster])

kmeans = KMeans(n_clusters=4, random_state=42)
df_scaled['Cluster'] = kmeans.fit_predict(df_scaled[cols_to_cluster])


product_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']


average_consumption_per_cluster = df_scaled.groupby('Cluster')[product_columns].mean()
print(average_consumption_per_cluster)

# Plotting
plt.figure(figsize=(16, 10))


ax = average_consumption_per_cluster.plot(kind='bar', figsize=(14, 8), edgecolor='black', colormap='Set3')
plt.title('Average Product Consumption per Cluster', fontsize=18, fontweight='bold')
plt.xlabel('Cluster', fontsize=14)
plt.ylabel('Average Consumption (Scaled)', fontsize=14)
plt.xticks(rotation=0, fontsize=12)


for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height() + 0.05),
                ha='center', va='bottom', fontsize=12)


plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.legend(title='Product', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)


plt.tight_layout()
plt.show()
```

```python

print("Missing values in 'Cluster':", df_scaled['Cluster'].isnull().sum())
print("Unique clusters:", df_scaled['Cluster'].unique())

df_scaled['Cluster'] = df_scaled['Cluster'].astype(int)


average_income_per_cluster = df_scaled.groupby('Cluster')['Income'].mean()

# Display the result
print("Average Income per Cluster:")
print(average_income_per_cluster)


plt.figure(figsize=(8, 5))
average_income_per_cluster.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Income per Cluster', fontsize=16, fontweight='bold')
plt.xlabel('Cluster', fontsize=12)
plt.ylabel('Average Income', fontsize=12)


for p in plt.gca().patches:
    plt.text(p.get_x() + p.get_width() / 2, p.get_height() + 0.05,
             round(p.get_height(), 2), ha='center', va='bottom', fontsize=12)


plt.xticks(rotation=0)


plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
